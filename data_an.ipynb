```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
# Order by price in descending order extractiong the top 10
df = df.sort_values(by='Price', ascending=False)
```

```python
# Create a brand phone lists
names = ['Apple', 'Samsung', 'Moto', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'OnePlus', 'Realme', 'Sony', 'LG']

# Create a empty list for each brand
name_lists = {name: 0 for name in names}

# Iterate over the rows of the dataframe
for index, row in df.iterrows():
    # Get the name of the phone
    name = row['Name'].split(' ')[0]
    # Check if the name is in the list
    if name in name_lists:
        # Increment the value of the brand
        name_lists[name] += 1

# Create a dataframe from the dictionary
df = pd.DataFrame(list(name_lists.items()), columns=['Brand', 'Count']).sort_values(by='Count', ascending=False)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Brand'], df['Count'])
plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Top 10 most expensive phones by brand')
plt.show()
```
